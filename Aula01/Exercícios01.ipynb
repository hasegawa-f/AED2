Projeto 1: A Agenda de Contatos:

• Contexto: Você precisa criar um sistema simples para armazenar e gerenciar contatos (nome,
telefone e e-mail).

• Problema: Como você pode adicionar novos contatos, buscar um contato pelo nome e listar
todos os contatos?

• Indicação de como resolver: Pensar em uma estrutura que permita associar informações
(telefone e e-mail) a um nome de forma eficiente.

• Resolução do Projeto 1: Um dicionário em Python é uma excelente escolha para este projeto.
A chave pode ser o nome do contato, e o valor pode ser outro dicionário contendo o telefone e
o e-mail.

```python
contatos = {'Ana': {'telefone': '1234-5678', 'e-mail': 'ana.dolores@email.com'},
            'Bruno': {'telefone': '9876-5432', 'e-mail': 'bruno_brunao@email.com'},
            'Carlos': {'telefone': '1122-3344', 'e-mail': 'carlos-engenheiro@email.com'}}

# Adicionar contato:
def adicionar_contato(nome, telefone, email):
    contatos[nome] =  {'telefone': telefone, 'e-mail': email}
    print(f'Novo contato adicionado:')
    print(f'Nome: {nome} \nTelefone: {telefone} \nE-mail: {email}\n')
    
# Buscar contato:
def buscar_contato(nome):
    if nome in contatos:
        print(f'''Contato encontrado:
        Nome: {nome}
        Telefone: {contatos[nome]['telefone']}
        E-mail: {contatos[nome]['e-mail']}\n''')
    else:
        print(f'Contato {nome} não encontrado\n')


# Listar contato (formatado):
def listar_contato():
    print('Lista de contatos:')
    for nome in contatos:
        print(f'Nome: {nome} \nTelefone: {contatos[nome]['telefone']} \nE-mail: {contatos[nome]['e-mail']}\n')
```

```python
adicionar_contato('Maria', '9876-5432', 'maria_contadora@email.com')
buscar_contato('Bruno')
buscar_contato('André')
listar_contato()
```

Projeto 2: A Lista de Tarefas:

• Contexto: Você precisa criar uma lista de tarefas para a semana.

• Problema: Como você pode adicionar novas tarefas, marcar tarefas como concluídas e
visualizar as tarefas pendentes?

• Indicação de como resolver: Pensar em uma forma de manter uma sequência de itens que
podem ser adicionados e removidos.

• Resolução do Projeto 2: Uma lista em Python é ideal para essa situação. Podemos adicionar
novas tarefas ao final da lista, remover tarefas quando concluídas e simplesmente imprimir a
lista para visualizar as tarefas pendentes.

```python
lista_de_tarefas = ['Lavar a louça', 'Estudar para a prova', 'Comprar ração para o gato', 'Limpar janela']

# Adicionar nova tarefa
def adicionar_tarefa(tarefa):
    lista_de_tarefas.append(tarefa)
    print(f'Nova tarefa adicionada: {tarefa}\n')

# Concluir uma tarefa
def concluir_tarefa(tarefa):
    if tarefa in lista_de_tarefas:
        lista_de_tarefas.remove(tarefa)
        print(f'Tarefa concluída e removida da lista: {tarefa}\n')
    else:
        print('Tarefa não encontrada na lista\n')

def listar_tarefas():
    print('Tarefas pendentes:')
    for tarefa in lista_de_tarefas:
        print(f'- {tarefa}')
    print()
```

```python
adicionar_tarefa('Tirar o lixo')
concluir_tarefa('Limpar janela')
concluir_tarefa('Guardar roupa')
listar_tarefas()
```
