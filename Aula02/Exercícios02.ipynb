## Exercício 3:
Qual a complexidade da busca binária? Considere que temos uma lista ordenada de tamanho n e
a função abaixo procura um elemento alvo nessa lista (retornando o índice ou -1 se não encontrar):

def busca_binaria(lista, alvo):

    esquerda = 0
    direita = len(lista) - 1
    while esquerda <= direita:
        meio = (esquerda + direita) // 2
        if lista[meio] == alvo:
            return meio
        elif lista[meio] < alvo:
            esquerda = meio + 1
        else:
            direita = meio - 1
    return -1
    
Qual é a ordem de complexidade em função de n?

#### Resposta:
Coomplexidade: O(log n)

Projeto 1: Análise de Desempenho de Busca em Dados de Clientes

Contexto: Imagine que a Fatec Votorantim possui um sistema com informações sobre seus ex-alunos. Cada
registro contém o nome, CPF e curso concluído. A coordenação precisa frequentemente buscar informações de
um aluno específico utilizando o CPF como identificador único. O sistema atual realiza uma busca sequencial
em uma lista de todos os ex-alunos.

Problema: Implemente duas funções em Python para buscar um ex-aluno pelo CPF em uma lista de dicionários
(onde cada dicionário representa um ex-aluno). A primeira função (buscar_aluno_linear) deve usar uma busca
linear, e a segunda função (buscar_aluno_eficiente) deve utilizar um dicionário para otimizar a busca,
assumindo que os dados podem ser pré-processados.

Indicação de como resolver: Para buscar_aluno_linear, itere pela lista de ex-alunos e compare o CPF de cada
um com o CPF buscado. Para buscar_aluno_eficiente, crie um dicionário onde a chave seja o CPF e o valor seja
o dicionário do ex-aluno correspondente. Analise a complexidade de tempo de ambas as abordagens.

```python
lista_de_alunos = [{'cpf': '123.456.789-10', 'nome': 'José Umdois', 'curso': 'Ciência de Dados Para Negócios'},
                  {'cpf': '321.654.987-01', 'nome': 'Roberto Trêsquestro', 'curso': 'Desenvolvimento de Software Multiplataforma'},
                  {'cpf': '231.564.897-11', 'nome': 'Larissa Cincoseis', 'curso': 'Controle de Obras'},
                  {'cpf': '132.465.798-00', 'nome': 'Alessandra Seteoito', 'curso': 'Ciência de Dados Para Negócios'},
                  {'cpf': '312.645.978-20', 'nome': 'Cláudio Novedez', 'curso': 'Controle de Obras'},
                  {'cpf': '213.546.879-02', 'nome': 'Cecília Onzedoze', 'curso': 'Desenvolvimento de Software Multiplataforma'}]

def buscar_aluno_linear(cpf, lista_alunos):
    for aluno in lista_alunos:
        if cpf == aluno['cpf']:
            return aluno
    return -1

def processar_lista(lista_alunos):
    nova_lista = {}
    for aluno in lista_alunos:
        cpf = aluno['cpf']
        nova_lista[cpf] = aluno
    return nova_lista

def buscar_aluno_eficiente(cpf, lista_alunos):
    if cpf in lista_alunos:
        return lista_alunos[cpf]
    else:
        return -1
```

```python
def buscar_aluno(aluno):
    if aluno != -1:
        print(f'''Aluno encontrado:
            CPF: {aluno['cpf']}
            Nome: {aluno['nome']}
            Curso: {aluno['curso']}\n''')
    else:
        print('Aluno não encontrado\n')

# Busca linear
aluno_buscado01 = buscar_aluno_linear('132.465.798-00', lista_de_alunos)
aluno_buscado02 = buscar_aluno_linear('111.111.111-11', lista_de_alunos)

buscar_aluno(aluno_buscado01)
buscar_aluno(aluno_buscado02)

# Busca eficiente
dicionario_alunos = processar_lista(lista_de_alunos)

aluno_buscado_e01 = buscar_aluno_eficiente('132.465.798-00', dicionario_alunos)
aluno_buscado_e02 = buscar_aluno_eficiente('111.111.111-11', dicionario_alunos)

buscar_aluno(aluno_buscado_e01)
buscar_aluno(aluno_buscado_e02)


# Complexidades: Lienar O(n) // Eficiente: O(1)
```

Projeto 2: Recomendação de Disciplinas Baseada em Interesses

Contexto: A Fatec Votorantim oferece diversas disciplinas optativas. Para ajudar os alunos a escolherem, a
coordenação quer criar um sistema simples de recomendação. Dado o histórico de disciplinas que um aluno já
cursou, o sistema deve sugerir outras disciplinas que alunos com interesses semelhantes também cursaram.

Problema: Dada uma lista de históricos de disciplinas de alunos (onde cada histórico é uma lista de códigos de
disciplinas) e o histórico de um aluno específico, implemente uma função em Python para encontrar disciplinas
que outros alunos com pelo menos uma disciplina em comum cursaram, mas que o aluno original ainda não
cursou.

Indicação de como resolver: Para cada outro aluno, compare seu histórico de disciplinas com o do aluno
original. Se houver alguma disciplina em comum, adicione as disciplinas que o outro aluno cursou e que o aluno
original não cursou a uma lista de recomendações. Analise a complexidade de tempo dessa abordagem.

```python
historicos = [
    ["ILP010", "BD001", "WEB001"],  
    ["MAT003", "FIS004"],           
    ["IHC002", "WEB001", "SEG005"], 
    ["ILP010", "IHC002"]
]

historico_aluno = ["ILP010", "IHC002"]

def recomendar_disciplinas(historico_alvo, lista_historicos):
    set_alvo = set(historico_alvo)
    recomendacoes = set() 
    
    for historico_outro in lista_historicos:
        set_outro = set(historico_outro)
        if set_alvo.intersection(set_outro): 
            materias_novas = set_outro.difference(set_alvo)
            recomendacoes.update(materias_novas)
    return list(recomendacoes)
```

```python
disciplinas_recomendadas = recomendar_disciplinas(historico_aluno, historicos)

print(disciplinas_recomendadas)

# A complexidade é O(N x M), sendo N o número total de alunos na lista_historicos e M o número máximo de disciplinas que um aluno cursou.
```
