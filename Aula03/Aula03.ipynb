```python
# Busca sequencial em vetor não ordenado

def busca_sequencial(vetor, chave):
    for i in vetor:
        if i == chave:
            return vetor.index(i)
    return -1

lista = [2, 78, 65, 34, 22, 9, 7, 14, 43, 75, 29]

busca_sequencial(lista, 99)
```

```python
#Resposta professor
def busca_seq_for(vetor, chave):
    n = len(vetor)
    for i in range(n):
        if vetor[i] == chave:
            return i

    return -1

def busca_seq_while(vetor, chave):
    n = len(vetor)
    i = 0
    while i < n:
        if vetor[i] == chave:
            return i
        i += 1
        
    return -1
```

```python
busca_seq_for(lista, 22)
```

```python
busca_seq_while(lista, 10)
```

```python
#Exercícios dos slides
# 1 - B
# 2 - A
# 3 - A
# 4:

def busca_cliente(nome, lista_clientes):
    n = len(lista_clientes)
    for i in range(n):
        if lista_clientes[i] == nome:
            return i
    return -1

lista_clientes = ["Ana", "João", "Maria", "Pedro", "Lucas", "Julia", "Mateus", "Mariana",
                  "Gabriel", "Laura", "Rafael", "Beatriz", "Carlos", "Fernanda", "Thiago"]

def verificar_presenca_cliente(cliente):
    indice_cliente = busca_cliente(cliente, lista_clientes)
    
    if indice_cliente != -1:
        print(f'O cliente informado se encontra na lista no índice {indice_cliente}')
    else:
        print('O cliente informado não se encontra na lista')
```

```python
verificar_presenca_cliente("Mariana")
```

```python
verificar_presenca_cliente("José")
```
