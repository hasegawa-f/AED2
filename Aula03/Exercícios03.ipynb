## Exercício 4
(PRÁTICA): Implemente uma função em Python chamada buscar_cliente(nome, lista_clientes) que
realiza busca sequencial em uma lista de nomes de clientes. Essa função deve retornar o índice onde o nome foi
encontrado ou -1 caso o nome não esteja na lista_clientes. Em seguida, escreva um pequeno código de
exemplo que utilize essa função para buscar por pelo menos um nome que exista e um que não exista na lista,
exibindo mensagens adequadas.

```python
def busca_cliente(nome, lista_clientes):
    n = len(lista_clientes)
    for i in range(n):
        if lista_clientes[i] == nome:
            return i
    return -1

lista_clientes = ["Ana", "João", "Maria", "Pedro", "Lucas", "Julia", "Mateus", "Mariana",
                  "Gabriel", "Laura", "Rafael", "Beatriz", "Carlos", "Fernanda", "Thiago"]

def verificar_presenca_cliente(cliente):
    indice_cliente = busca_cliente(cliente, lista_clientes)
    
    if indice_cliente != -1:
        print(f'O cliente informado se encontra na lista no índice {indice_cliente}\n')
    else:
        print('O cliente informado não se encontra na lista\n')
```

```python
verificar_presenca_cliente("Mariana")
```

```python
verificar_presenca_cliente("José")
```

## Exercício 5
(PRÁTICA): Crie uma função conta_ocorrencias(valor, lista) em Python que percorre uma lista e conta
quantas vezes um determinado valor aparece. Utilize busca sequencial (varrendo a lista) para realizar a
contagem. Por exemplo, conta_ocorrencias(3, [1,3,5,3,7]) deve retornar 2, pois o valor 3 aparece duas
vezes na lista. Inclua casos de teste chamando a função com diferentes valores e listas.


```python
def conta_ocorrencias(valor, lista):
    ocorrencias = 0
    for i in range(len(lista)):
        if valor == lista[i]:
            ocorrencias += 1
    return ocorrencias
```

```python
lista_01 = [7, 3, 5, 7, 1, 9, 3, 7, 2, 5, 8, 7, 3, 1]
lista_02 = ['A', 'B', 'C', 'A', 'D', 'B', 'E', 'A', 'C', 'F', 'A', 'B', 'C']
lista_03 = [2, 2, 4, 1, 3, 2, 4, 1, 1, 5, 2, 3, 1, 4, 2]

ocorrencias_01 = conta_ocorrencias(7, lista_01)
ocorrencias_02 = conta_ocorrencias('B', lista_02)
ocorrencias_03 = conta_ocorrencias(3, lista_03)

print(ocorrencias_01)
print(ocorrencias_02)
print(ocorrencias_03)
```

## Exercício 6
(PRÁTICA): Suponha que precisamos remover um elemento específico de uma lista não ordenada. Escreva
uma função Python remove_elemento(lista, valor) que remove a primeira ocorrência de valor na lista,
mantendo os demais elementos e retornando o valor removido. A função deve retornar None caso o valor não
seja encontrado ou a lista esteja vazia. Dica: Utilize a busca sequencial para encontrar a posição do elemento
e, se encontrada, deslize os elementos posteriores para “preencher” o espaço removido (ou use métodos
próprios da lista Python). Demonstre o funcionamento da função com um exemplo

```python
def remove_elemento(lista, valor):
    for i in range(len(lista)):
        if valor == lista[i]:
            return lista.pop(i)
    return None
```

```python
lista_numeros = [2, 2, 4, 1, 3, 2, 4, 1, 1, 5, 2, 3, 1, 4, 2]

print(f' Lista original: {lista_numeros}')
elemento_removido = remove_elemento(lista_numeros, 5)
print(f'Elemento removido: {elemento_removido}')
print(f'Lista atualizada: {lista_numeros}')
```

## Projeto 1: Encontrando seu Filme Favorito  
Contexto: Imagine um sistema de recomendação de filmes simples onde os filmes são armazenados em uma
lista não-ordenada.

Problema: Implementar uma função que permita ao usuário buscar um filme pelo título nessa lista.

Indicação de Solução: Crie uma lista de títulos de filmes e uma função que use a busca sequencial para
encontrar um título inserido pelo usuário. Se o filme for encontrado, exiba uma mensagem informando o título e
sua posição (índice) na lista. Se não for encontrado, exiba uma mensagem informando que o filme não está no
catálogo.


```python
lista_filmes = [
    "Parasita",
    "Mad Max: Estrada da Fúria",
    "O Auto da Compadecida",
    "A Viagem de Chihiro",
    "Entre Facas e Segredos",
    "Matrix",
    "Homem-Aranha no Aranhaverso"
]

def busca_filme(lista):
    filme_buscado = input('Digite o nome do filme desejado')
    for i in range(len(lista)):
        if filme_buscado == lista[i]:
            print(f'Filme {filme_buscado} encontrado na posição {i}')
            return
    print(f'Filme {filme_buscado} não encontrado na lista')
```

```python
busca_filme(lista_filmes)
```

```python
busca_filme(lista_filmes)
```

## Projeto 2: Verificando a Lista de Convidados
Contexto: Desenvolver um programa para gerenciar a lista de convidados de um evento.

Problema: Implementar uma funcionalidade que permita verificar se um nome específico está na lista de
convidados.

Indicação de Solução: Crie uma lista de nomes de convidados. Implemente uma função que receba um nome
como entrada e use a busca sequencial para verificar se esse nome existe na lista. Exiba uma mensagem
indicando se o convidado está na lista ou não

```python
lista_convidados = ["Tiago", "Ana", "João", "Sofia", "Miguel", "Rita", "Pedro", "Inês", "Rui", "Maria"]

def busca_convidado(lista):
    convidado = input('Digite o nome do convidado')
    for i in range(len(lista)):
        if convidado == lista[i]:
            print(f'Convidado(a) {convidado} está na lista!')
            return
    print(f'Convidado(a) {convidado} não está presente na lista!')
```

```python
busca_convidado(lista_convidados)
```

```python
busca_convidado(lista_convidados)
```
