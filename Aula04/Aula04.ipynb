Busca Eficiente em um Índice de Livros

```python
def busca_livros(titulo_livro, lista_livros):
    inicio = 0
    fim = len(lista_livros) - 1
    while inicio <= fim:
        meio = (inicio + fim) // 2
        if titulo_livro == lista_livros[meio][0]:
            return lista_livros[meio]
        elif titulo_livro > lista_livros[meio][0]:
            inicio = meio + 1
        else:
            fim = meio - 1
    return None
```
