## Exercício 1
(ENADE): A busca binária possui complexidade de tempo diferente da busca sequencial. Suponha um vetor
de tamanho n já ordenado. Qual das alternativas descreve corretamente a complexidade de tempo da busca
binária, em termos de melhor caso e pior caso, comparando-a à busca sequencial?  
A. Melhor caso O(n) e pior caso O(n²) na busca binária, o que é mais eficiente que a busca sequencial.  
B. Melhor caso O(1) e pior caso O(log n) na busca binária, enquanto a busca sequencial tem melhor caso O(1)  
e pior caso O(n).  
C. Melhor caso e pior caso O(log n) na busca binária, pois sempre dividimos a busca pela metade; a busca  
sequencial sempre é O(n) em qualquer caso.  
D. Melhor caso O(1) e pior caso O(n) tanto para busca binária quanto para busca sequencial.

#### Resposta: B. Melhor caso O(1) e pior caso O(log n) na busca binária, enquanto a busca sequencial tem melhor caso O(1) e pior caso O(n).

## Exercício 2
(TEÓRICA): Por que o algoritmo de busca binária exige que o vetor esteja ordenado previamente? O que
poderia acontecer se tentássemos aplicar a busca binária em um conjunto de dados não ordenado?

#### Resposta:  
Porque, dado que o valor não foi encontrado na primeira iteração, o vetor será dividido ao meio e esse será utilizado para a próxima iteração.
Se o valor que está no índice do meio não for, de fato, o meio, o valor procurado pode ficar na parte do vetor que foi "ignorada" na próxima iteração e, assim, a busca não será bem-sucedida

## Exercício 3
(PRÁTICA): Implemente a função busca_binaria de forma recursiva em Python. Ou seja, escreva uma versão
da busca binária que chame a si própria recursivamente em vez de usar um loop while. A função deve receber
como parâmetros o vetor ordenado, o valor alvo buscado, e opcionalmente os índices esq e dir delimitando o
segmento de busca (ou você pode definir a função interna aninhada com esses parâmetros). Retorne o índice
do alvo no vetor caso seja encontrado, ou -1 caso contrário.

```python
def busca_binaria_recursiva(numero, lista, inicio, fim = None):
    if fim is None:
        fim = len(lista) - 1
    if inicio > fim:
        return -1
    meio = (inicio + fim) // 2
    if numero == lista[meio]:
        return meio
    elif numero > lista[meio]:
        inicio = meio + 1
        return busca_binaria_recursiva(numero, lista, inicio, fim)
    elif numero < lista[meio]:
        fim = meio - 1
        return busca_binaria_recursiva(numero, lista, inicio, fim)
```

```python
lista_numeros_ord = range(1, 100)
```

```python
indice = busca_binaria_recursiva(56, lista_numeros_ord, 0)
print(f'Indice do numero encontrado: {indice}')
```

## Exercício 4
(PRÁTICA): Considere uma lista Python de 1000 números inteiros já ordenados em ordem crescente. Escreva
um pequeno programa que utilize tanto a busca sequencial quanto a busca binária para encontrar um
determinado número nessa lista e conte o número de comparações feitas em cada método. Faça o teste com
um número que não esteja presente na lista (para forçar o pior caso) e exiba na saída o total de comparações
realizadas por cada método. (Dica: você pode modificar as funções de busca para incrementarem um contador
global ou utilizar variáveis externas para contar as comparações.)

```python
def busca_sequencial_numeros(numero, lista):
    comparacoes = 0
    for i in range(len(lista)):
        comparacoes += 1
        if numero == lista[i]:
            return comparacoes
    return comparacoes

def busca_binaria_numero(numero, lista):
    comparacoes = 0
    inicio = 0
    fim = len(lista) - 1
    while inicio <= fim:
        comparacoes += 1
        meio = (inicio + fim) // 2
        if numero == lista[meio]:
            return comparacoes
        elif numero > lista[meio]:
            inicio = meio + 1
        else:
            fim = meio - 1
    return comparacoes
```

```python
lista_numeros = list(range(1, 1001))

comparacoes_01 = busca_sequencial_numeros(457, lista_numeros)
comparacoes_02 = busca_sequencial_numeros(1000, lista_numeros)
comparacoes_03 = busca_binaria_numero(457, lista_numeros)
comparacoes_04 = busca_binaria_numero(1000, lista_numeros)

print(f'Primeiro caso sequencial, buscando 457: {comparacoes_01} comparações')
print(f'Pior caso sequencial, buscando 1000: {comparacoes_02} comparações')
print(f'Primeiro caso binaria, buscando 457: {comparacoes_03} comparações')
print(f'Pior caso binaria, buscando 1000: {comparacoes_04} comparações')
```

## Projeto 1: Busca Eficiente em um Catálogo de Produtos Online  
Contexto: Imagine que você está desenvolvendo a funcionalidade de busca para um site de e-commerce com
um grande número de produtos. O catálogo de produtos é armazenado em uma lista, onde cada item é um
dicionário contendo informações como nome, preço e descrição. Para otimizar a busca, o catálogo é mantido
ordenado alfabeticamente pelo nome do produto.  

Problema: Os clientes precisam encontrar rapidamente informações sobre um produto específico digitando
seu nome na barra de busca. Uma busca lenta pode levar à frustração e à perda de vendas.

Indicação de como resolver: Implemente um sistema de busca que utilize a busca binária para localizar o
produto no catálogo ordenado. Ao receber o nome do produto digitado pelo usuário, sua função deve realizar a
busca binária na lista de produtos. Se o produto for encontrado, retorne suas informações (por exemplo, preço,
descrição). Caso contrário, informe que o produto não foi encontrado.


```python
def busca_produtos(nome_produto, lista_produtos):
    inicio = 0
    fim = len(lista_produtos) - 1
    while inicio <= fim:
        meio = (inicio + fim) // 2
        if nome_produto == lista_produtos[meio]['nome']:
            return lista_produtos[meio]
        elif nome_produto > lista_produtos[meio]['nome']:
            inicio = meio + 1
        else:
            fim = meio - 1
    return None
```

## Projeto 2: Busca Eficiente em um Índice de Livros  
Contexto: Você está trabalhando em um projeto para gerenciar uma biblioteca. As informações sobre os livros
(título, autor, ano de publicação) estão armazenadas em uma lista, e essa lista é mantida ordenada
alfabeticamente pelo título do livro.

Problema: Os usuários precisam encontrar rapidamente um livro específico sabendo seu título. Uma busca
ineficiente pode dificultar a localização dos livros desejados.

Indicação de como resolver: Desenvolva uma funcionalidade que permita aos usuários buscar um livro pelo
título utilizando a busca binária. Ao receber o título do livro, sua função deve procurar no índice ordenado. Se o
livro for encontrado, exiba suas informações (título, autor, ano). Caso contrário, informe que o livro não foi
encontrado.

```python
livros_python = [
    ("Aprenda Python 3 do Jeito Difícil", "Zed Shaw", 2017),
    ("Arquitetura Limpa", "Robert C. Martin", 2017),
    ("Automate the Boring Stuff with Python", "Al Sweigart", 2015),
    ("Código Limpo (Clean Code)", "Robert C. Martin", 2008),
    ("Expressões Regulares: Uma Abordagem Divertida", "Aurelio Marinho Jargas", 2012),
    ("Grokking Algorithms", "Aditya Bhargava", 2016),
    ("Introdução à Programação com Python", "Nilo Ney Coutinho Menezes", 2010),
    ("Pense em Python", "Allen B. Downey", 2012),
    ("Python Fluente", "Luciano Ramalho", 2015),
    ("Python Intensivo", "Eric Matthes", 2015)
]

def busca_livros(titulo_livro, lista_livros):
    inicio = 0
    fim = len(lista_livros) - 1
    while inicio <= fim:
        meio = (inicio + fim) // 2
        if titulo_livro == lista_livros[meio][0]:
            return lista_livros[meio]
        elif titulo_livro > lista_livros[meio][0]:
            inicio = meio + 1
        else:
            fim = meio - 1
    return None
```

```python
def mensagem_livro(livro_buscado):
    if livro_buscado != None:
        print(f'''Livro encontrado!
        Título: {livro_buscado[0]}
        Autor(a): {livro_buscado[1]}
        Ano de Publicação: {livro_buscado[2]}''')
    else:
        print('Livro não encontrado!')
```

```python
livro_buscado = busca_livros("Pense em Python", livros_python)
mensagem_livro(livro_buscado)
```

```python
livro_buscado = busca_livros("Clean Code", livros_python)
mensagem_livro(livro_buscado)
```
