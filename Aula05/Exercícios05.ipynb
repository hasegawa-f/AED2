## Exercício 5
(PRÁTICA): Classe Pessoa: Crie uma classe Pessoa com nome e idade. Adicione método
faz_aniversario() que incrementa a idade em 1. Use `__str__()` para retornar algo como "`<nome>`, `<idade>`
anos". Crie duas pessoas e mostre a idade de cada uma antes e depois de chamar faz_aniversario()

````python
class Pessoa:
    def __init__(self, nome, idade):
        self.nome = nome
        self.idade = idade

    def __str__(self):
        return f'Nome: {self.nome} \nIdade: {self.idade}\n'

    def faz_aniversario(self):
        self.idade += 1
````

````python
pessoa_01 = Pessoa('Alfredo', 55)
pessoa_02 = Pessoa('Josi', 43)

print('Antes do aniversário')
print(pessoa_01)
print(pessoa_02)

pessoa_01.faz_aniversario()
pessoa_02.faz_aniversario()

print('Após o aniversário')
print(pessoa_01)
print(pessoa_02)
````

## Projeto 1 Projeto 1: Biblioteca de Livros

__Contexto do mundo real__: Imagine um sistema para gerenciar o empréstimo de livros em uma pequena
biblioteca escolar. Queremos cadastrar livros e usuários, registrar quando alguém empresta ou devolve um
livro.

__Descrição do problema__: Cada Livro tem título, autor, quantidade disponível. Cada Usuário tem nome,
matrícula e lista de livros emprestados. O sistema deve permitir que um usuário faça empréstimos (reduzindo
a quantidade disponível) e devoluções (aumentando quantidade). Devemos evitar empréstimos se o livro
estiver indisponível. Também precisamos manter os dados organizados e seguros (ex.: não permitir mexer
diretamente na quantidade disponível).


````python
class Livro:
    def __init__(self, titulo, autor, qtd_disponivel):
        self.__titulo = titulo
        self.__autor = autor
        self.__qtd_disponivel = qtd_disponivel

    def __str__(self):
        return f'Titulo: {self.__titulo} \nAutor: {self.__autor}'
    
    def get_titulo(self):
        return self.__titulo

    def get_autor(self):
        return self.__autor
    
    def get_qtd(self):
        return self.__qtd_disponivel

    def emprestimo(self):
        self.__qtd_disponivel -= 1

    def retorno(self):
        self.__qtd_disponivel += 1
````

````python
class Usuario:
    def __init__(self, nome, matricula, lista_livros):
        self.__nome = nome
        self.__matricula = matricula
        self.__lista_livros = []

    def get_nome(self):
        return self.__nome

    def get_matricula(self):
        return self.__matricula

    def get_lista_livros(self):
        return self.__lista_livros
    
    def get_lista_titulos(self):
        if len(self.__lista_livros) == 0:
            return []
        else:
            lista_titulos = []
            for livro in self.__lista_livros:
                lista_titulos.append(livro.get_titulo())
            return lista_titulos

    def emprestar_livro(self, livro):
        self.__lista_livros.append(livro)

    def devolver_livro(self, livro):
        self.__lista_livros.remove(livro)
````

````python
livro_01 = Livro('O Senhor dos Anéis', 'J.R.R. Tolkien', 3)
livro_02 = Livro('1984', 'George Orwell', 5)
livro_03 = Livro('Dom Quixote', 'Miguel de Cervantes', 2)

usuario_01 = Usuario('Lucas', 48291, [])
usuario_02 = Usuario('Camila', 73504, [])
usuario_03 = Usuario('Bruno', 15937, [])
````

````python
def emprestimo(usuario, novo_livro):
    if novo_livro.get_qtd() == 0:
        print('Livro não disponível!')
        return
    lista_livros_usuario = usuario.get_lista_livros()
    for livro in lista_livros_usuario:
        if novo_livro.get_titulo() == livro.get_titulo():
            print('Livro já consta na lista do usuário!\n')
            return
    usuario.emprestar_livro(novo_livro)
    novo_livro.emprestimo()
    print(f'Livro "{novo_livro.get_titulo()}" emprestado com sucesso para o usuario {usuario.get_nome()}, matricula {usuario.get_matricula()}')
    print(f'Quantidade disponível para empréstimo: {novo_livro.get_qtd()}')
    return

def devolucao(usuario, livro_devolvido):
    lista_livros_usuario = usuario.get_lista_livros()
    if livro_devolvido not in lista_livros_usuario:
        print('Erro! Usuario não contém livro em seu registro de empréstimos! Cancelando')
        return
    usuario.devolver_livro(livro_devolvido)
    livro_devolvido.retorno()
    print(f'Livro "{livro_devolvido.get_titulo()}" devolvido com sucesso pelo usuário {usuario.get_nome()}, matricula {usuario.get_matricula()}')
    print(f'Quantidade disponível para empréstimo: {livro_devolvido.get_qtd()}')
````

__Contexto do mundo real__: Um banco que oferece dois tipos de conta: conta corrente e conta poupança.
Queremos um sistema que permita depositar, sacar e aplicar juros (no caso da poupança).

__Descrição do problema__: Precisamos modelar contas bancárias em Python de forma orientada a objetos.
Teremos:

- Classe base Conta com atributos comuns (titular, saldo) e métodos depositar(valor) e sacar(valor).
- Classe ContaCorrente que herda de Conta e adiciona, por exemplo, taxa de saque (ou limite
especial).
- Classe ContaPoupanca que herda de Conta e tem um método extra render_juros() que aumenta o
saldo de acordo com uma taxa fixa anual.
Também podemos manter atributos de classe (como juros = 0.05 para 5% ao ano).

````python
class Conta:
    def __init__(self, titular, saldo):
        self._titular = titular
        self._saldo = saldo

    def __str__(self):
        return f'Titular da conta: {self._titular} \nSaldo Disponível: {self._saldo:.2f}'

    def get_titular(self):
        return self._titular

    def get_saldo(self):
        return self._saldo
    
    def depositar_valor(self, valor):
        self._saldo += valor

    def subtrair_saldo(self, valor):
        self._saldo -= valor

    def sacar_valor(self, valor):
        if self._saldo - valor < 0:
            return False
        else:
            self.subtrair_saldo(valor)
            return True
````

````python
class Conta_Corrente(Conta):
    def __init__(self, titular, saldo, limite_especial):
        super().__init__(titular, saldo)
        self.__limite_especial = limite_especial
        self.__limite_especial_atual = limite_especial

    def get_limite_especial(self):
        return self.__limite_especial
    
    def get_limite_especial_atual(self):
        return self.__limite_especial_atual

    def subtrair_limite_especial(self, valor):
        self.__limite_especial_atual -= valor

    def saque_especial(self, valor):
        if self.__limite_especial_atual - valor < 0:
            return False
        else:
            self.subtrair_limite_especial(valor)
            return True

    def restaurar_limite_especial(self, valor):
        if self.__limite_especial_atual + valor > self.__limite_especial:
            self.__limite_especial_atual = self.__limite_especial
        else:
            self.__limite_especial_atual += valor
````

````python
class Conta_Poupanca(Conta):
    juros = 0.05
    
    def __init__(self, titular, saldo):
        super().__init__(titular, saldo)
    
    def render_juros(self):
        self._saldo *= juros
````
