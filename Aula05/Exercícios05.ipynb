## Exercício 3
(TEÓRICA): Comparando paradigmas: Qual das opções melhor diferencia a programação procedural da
programação orientada a objetos?  
A. Na programação procedural, usa-se funções; na orientada a objetos, não existem funções.  
B. A programação orientada a objetos organiza o código em classes/objetos que combinam estado e
comportamento, enquanto a procedural organiza o código em rotinas (funções) separadas dos dados.  
C. A programação procedural é sempre menos eficiente que a orientada a objetos.  
D. A orientação a objetos não permite estrutura de controle como loops e condicionais, ao contrário do
procedural.

## Exercício 4

instâncias.

## Projeto 1 Projeto 1: Biblioteca de Livros

__Contexto do mundo real__: Imagine um sistema para gerenciar o empréstimo de livros em uma pequena
biblioteca escolar. Queremos cadastrar livros e usuários, registrar quando alguém empresta ou devolve um
livro.

__Descrição do problema__: Cada Livro tem título, autor, quantidade disponível. Cada Usuário tem nome,
matrícula e lista de livros emprestados. O sistema deve permitir que um usuário faça empréstimos (reduzindo
a quantidade disponível) e devoluções (aumentando quantidade). Devemos evitar empréstimos se o livro
estiver indisponível. Também precisamos manter os dados organizados e seguros (ex.: não permitir mexer
diretamente na quantidade disponível).


## Projeto 2 Gerenciamento de Conta Bancária com Herança

__Contexto do mundo real__: Um banco que oferece dois tipos de conta: conta corrente e conta poupança.
Queremos um sistema que permita depositar, sacar e aplicar juros (no caso da poupança).

__Descrição do problema__: Precisamos modelar contas bancárias em Python de forma orientada a objetos.
Teremos:

- Classe base Conta com atributos comuns (titular, saldo) e métodos depositar(valor) e sacar(valor).
- Classe ContaCorrente que herda de Conta e adiciona, por exemplo, taxa de saque (ou limite
especial).
- Classe ContaPoupanca que herda de Conta e tem um método extra render_juros() que aumenta o
saldo de acordo com uma taxa fixa anual.
Também podemos manter atributos de classe (como juros = 0.05 para 5% ao ano).
