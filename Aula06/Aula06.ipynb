```python
class Stack:
    def __init__(self):
        self.itens = []
    
    def is_empty(self):
        return self.itens == []

    def print_numero_convertido(self):
        binario = ''
        while not self.is_empty():
            binario = binario + str(self.pop())
        print(f'binario -> {binario}')

    def converter_numero(self, numero_d):
        print(f'decimal -> {numero_d}')
        while numero_d > 0:
            digito_b = numero_d % 2
            self.push(digito_b)
            numero_d //= 2
        self.print_numero_convertido()
    
    def push(self, item):
        self.itens.append(item)

    def pop(self):
        return self.itens.pop()
    
    def peek(self):
        return self.itens[-1]
    
    def size(self):
        return len(self.itens)
    
    def print_stack(self):
        print(self.itens)
```

```python
pilha = Stack()
```

```python
pilha.converter_numero(233)
```

```python
pilha.print_stack()
```

```python
pilha.converter_numero(352)
```
