## Exercício 05
(Prática) Implemente em Python uma função decimal_para_hexa(n) que converta um número decimal nãonegativo n para sua representação hexadecimal como string, usando uma pilha. (Por exemplo, decimal_para_hexa(26) deve retornar "1A".)


````python
def converter_digito(d):
    match d:
        case 10:
             return 'A'
        
        case 11:
             return 'B'
        
        case 12:
             return 'C'
    
        case 13:
             return 'D'
        
        case 14:
             return 'E'
        
        case 15:
             return 'F'
        
        case _:
            return d

def print_numero_convertido(pilha):
    hexa_d = ''
    while len(pilha) > 0:
        hexa_d += str(pilha.pop())
    print(f'hexadecimal: {hexa_d}')

def decimal_para_hexadecimal(n):
    print(f'decimal: {n}')
    pilha = []
    while n > 0:
        digito = n % 16
        digito = converter_digito(digito)
        pilha.append(digito)
        n //= 16
    print_numero_convertido(pilha)
        
````

````python
decimal_para_hexadecimal(26)
````
