```python
class Queue:
    def __init__(self):
        self.itens = []
        
    def is_empty(self):
        return self.itens == []
        
    def enqueue(self, item):
        self.itens.insert(0, item)
        
    def dequeue(self):
        return self.itens.pop()
        
    def size(self):
        return len(self.itens)
        
    def print_queue(self):
        print(self.itens)
```

## Exercício 1: 

Jogo: Batata Quente

- Neste jogo, as crianças formam um círculo e passam um item qualquer (batata) cada
um para o seu vizinho da frente o mais rápido possível.
- Em um certo momento do jogo, essa ação é interrompida (queimou) e a criança que
estiver com o item (batata) na mão é excluída da roda.
- O jogo então prossegue até que reste apenas uma única criança, que é a vencedora.
- Para simular um círculo (roda), utilizaremos uma fila da seguinte maneira: a criança
que está com a batata na mão será sempre aquela que estiver no início da fila. Após
passar a batata, a simulação deve instantaneamente remover e inserir a criança,
colocando-a novamente no final da fila. Ela então vai esperar até que todas as outras
assumam o início da fila, antes de assumir essa posição novamente.
- Após um número pré estabelecido MAX de operações enqueue/dequeue, a criança
que ocupar o início da fila será removida e outro ciclo da brincadeira é realizado.
- O processo continua até que a fila tenha possua tamanho um. 

```python
from random import randint

def batata_quente(fila):
    while fila.size() > 1:
        rodadas = randint(fila.size()//2, fila.size()*2)
        for i in range(rodadas+1):
            jogador = fila.dequeue()
            fila.enqueue(jogador)
        jogador_queimado = fila.dequeue()    
        print(f'Queimou! Jogador eliminado: {jogador_queimado}')

    print(f'\nFim de jogo! Vencedor: {fila.itens[0]}')
```

```python
nomes = ['Miguel', 'Helena', 'Arthur', 'Alice', 'Gael', 
         'Laura', 'Heitor', 'Maria Alice', 'Theo', 'Sophia', 
         'Davi', 'Manuela', 'Gabriel', 'Maitê', 'Noah', 'Cecília',
         'Samuel', 'Isabella', 'Pedro', 'Luísa', 'Benício', 'Valentina',
         'Isaac', 'Elisa', 'Anthony']

fila = Queue()
fila.itens = nomes
```

```python
batata_quente(fila)
```

## Exercício 5

Atendimento em um banco:
 
- Simule uma fila de atendimento em um banco, onde as pessoas são chamadas na ordem em que chegaram. O próximo cliente da fila é atendido e removido da fila. A cada três atendimentos, o atendente faz uma pausa, e novos clientes podem ser adicionados à fila durante esse período.
- __Requisitos__:  
1. Criar a fila com os nomes de clientes.  
2. Processar três atendimentos e depois fazer uma pausa.  
3. Durante a pausa, adicionar dois novos clientes à fila.  
4. Continuar até que todos os clientes tenham sido atendidos.

```python
def atendimento_banco(lista_espera):
    clientes = Queue()
    for i in range(10):
        novo_cliente = lista_espera.pop()
        clientes.enqueue(novo_cliente)
        print(f'Novo cliente na fila: {novo_cliente}')
    while not clientes.is_empty():
        print('\nInício do atendimento!\n')
        if clientes.size() >= 3:
            numero_atendimentos = 3
        elif clientes.size() == 2:
            numero_atendimentos = 2
        elif clientes.size() == 1:
            numero_atendimentos = 1
            
        for i in range(numero_atendimentos):
            cliente_atendido = clientes.dequeue()
            print(f'Cliente atendido: {cliente_atendido}')
            
        print('\nPausa no atendimento!\n')
        if len(lista_espera) >= 2:
            for i in range(2):
                novo_cliente = lista_espera.pop()
                clientes.enqueue(novo_cliente)
                print(f'Novo cliente na fila: {novo_cliente}')
        elif len(lista_espera) == 1:
            novo_cliente = lista_espera.pop()
            clientes.enqueue(novo_cliente)
            print(f'Novo cliente na fila: {novo_cliente}')
            
    print('Não há mais clientes na fila')    
    
```

```python
lista_clientes = ['Ricardo', 'Fernanda', 'Alessandro', 'Beatriz', 
                'Roberto', 'Luciana', 'Juliana', 'Marcelo',
                'Cláudia', 'Rodrigo', 'Patrícia', 'Maurício',
                'Tatiana', 'Gustavo', 'Danielle']

atendimento_banco(lista_clientes)
```
