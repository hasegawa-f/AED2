## Exercício 3

(TEORIA). Em uma lista encadeada simples com n nós, qual é a complexidade de tempo da operação de busca (search)?

A. O(1)  
B. O(log n)  
C. O(n)  
D. O(n²)  
E. O(n log n)

#### Resposta: C. O(n)

```python
class Node:
    def __init__(self, init_data):
        self.data = init_data
        self.next = None
    
    def get_data(self):
        return self.data
    
    def get_next(self):
        return self.next

    def set_data(self, new_data):
        self.data = new_data

    def set_next(self, new_next):
        self.next = new_next
```

```python
class List:
    def __init__(self):
        self.head = None

    def is_empty(self):
        return self.head == None
    
    def add_head(self, item):
        temp = Node(item)
        temp.set_next(self.head)
        self.head = temp
    
    def add_tail(self, item):
        tail = Node(item)
        temp = self.head
        while temp.next != None:
            temp = temp.next
        temp.set_next(tail)
    
    def print_list(self):
        temp = self.head
        X = []
        while temp != None:
            X.append(temp.data)
            temp = temp.get_next()
        return X
    
    def size(self):
        temp = self.head
        count = 0
        while temp != None:
            count = count + 1
            temp = temp.get_next()
        return count
    
    def search(self,item):
        temp = self.head
        found = False
        while temp != None and not found:
            if temp.get_data() == item:
                found = True
            else:
                temp = temp.get_next()
        return found
    
    def remove(self, item):
        current = self.head
        previous = None
        found = False
        while not found:
            if current.get_data() == item:
                found = True
            else:
                previous = current
                current = current.get_next()
        if previous == None:
            self.head = current.get_next()
        else:
            previous.set_next(current.get_next())
        current.set_next(None)
```

## Exercício Prático 1

Escreva uma função para encontrar o comprimento de uma lista encadeada simples.  
Chame-a de: get_length() e como parâmetro deve ser passada a lista encadeada.

```python
def get_length(linked_list):
    temp = linked_list.head
    if temp == None:
        return 0
    count = 1
    while temp.get_next() != None:
        count += 1
        temp = temp.get_next()
    return count
    
```

```python
no01 = Node(1)
no02 = Node(2)
no03 = Node(3)
no04 = Node(4)
no05 = Node(5)

link_list = List()

print(get_length(link_list))

link_list.add_head(no01)
link_list.add_tail(no02)
link_list.add_tail(no03)
link_list.add_tail(no04)
link_list.add_tail(no05)
```

## Exercício Prático 2: Histórico de Navegação

Implemente uma lista encadeada que simula o histórico de navegação de um navegador. A lista deve permitir:  
1.Adicionar uma nova página ao histórico.  
2.Exibir o histórico completo.  
3.Voltar uma página no histórico (remover o último elemento).

```python
class Historico_Navegacao:
    def __init__(self):
        self.head = None

    def is_empty(self):
        return self.head == None
    
    def add_pagina(self, pagina):
        if self.head == None:
            self.head = pagina
        else:
            tail = pagina
            temp = self.head
            while temp.next != None:
                temp = temp.next
            temp.set_next(tail)
    
    def exibir_historico(self):
        temp = self.head
        X = []
        while temp != None:
            X.append(temp.data)
            temp = temp.get_next()
        return X
    
    def remove_ultima_pagina(self):
        if self.head == None:
            return
        elif self.head.get_next() == None:
            self.head = None
        else: 
            temp = self.head
            while temp.next != None:
                temp02 = temp
                temp = temp.get_next()
            temp02.next = None
        
```

```python
pagina01 = Node('Receitas Deliciosas')
pagina02 = Node('Filmes e Séries')
pagina03 = Node('Jogos Online')
pagina04 = Node('Procurar')

historico = Historico_Navegacao()

historico.add_pagina(pagina01)
historico.add_pagina(pagina02)
historico.add_pagina(pagina03)
historico.add_pagina(pagina04)
```

## Exercício Prático 3: Lista de Espera em um Consultório

Implemente uma lista encadeada para gerenciar uma lista de espera em um consultório médico. O sistema deve permitir:  
1.Adicionar um paciente ao final da lista de espera.  
2.Chamar o próximo paciente (remover o primeiro da lista).  
3.Exibir a lista de espera.

```python
class Fila_Consultorio:
    def __init__(self):
        self.head = None

    def is_empty(self):
        return self.head == None
    
    def add_paciente(self, paciente):
        if self.head == None:
            self.head = paciente
        else:
            tail = paciente
            temp = self.head
            while temp.next != None:
                temp = temp.next
            temp.set_next(tail)
    
    def exibir_fila(self):
        temp = self.head
        X = []
        while temp != None:
            X.append(temp.data)
            temp = temp.get_next()
        return X
    
    def atender_paciente(self):
        if self.head == None:
            print('Nenhum paciente na fila')
        elif self.head.get_next() == None:
            self.head = None
        else: 
            self.head = self.head.get_next()
```

```python
paciente01 = Node('Ester')
paciente02 = Node('Ramon')
paciente03 = Node('Julia')
paciente04 = Node('Caio')
paciente05 = Node('Olivia')

fila = Fila_Consultorio()

fila.add_paciente(paciente01)
fila.add_paciente(paciente02)
fila.add_paciente(paciente03)
fila.add_paciente(paciente04)
fila.add_paciente(paciente05)
```

```python
fila.exibir_fila()
```

```python
fila.atender_paciente()
```

```python
fila.exibir_fila()
```
