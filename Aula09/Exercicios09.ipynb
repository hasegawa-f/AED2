## Exercício Prático 1

Escreva uma função recursiva em Python chamada soma_digitos(n) que calcule a soma dos dígitos de um número inteiro positivo n. Por exemplo, soma_digitos(123) deve retornar 1 + 2 + 3 = 6.

```python
def soma_digitos(dig):
    soma = 0
    if dig // 10 == 0:
        soma += dig
        return soma
    else:
        soma += dig % 10
        return soma + soma_digitos(dig//10)
```

```python
soma_digitos(123)
```

```python
soma_digitos(123456789)
```
